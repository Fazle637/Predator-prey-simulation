```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
a = 0.6   # prey birth rate
b = 0.025 # predation rate
c = 0.4   # predator death rate
d = 0.015 # predator growth rate

# Time settings
h = 0.01          # time step
t_max = 50
time = np.arange(0, t_max, h)

# Initial populations
prey = np.zeros(len(time))
predator = np.zeros(len(time))

prey[0] = 40      # initial prey
predator[0] = 9   # initial predator

# Euler Method
for i in range(1, len(time)):
    prey[i] = prey[i-1] + h * (a*prey[i-1] - b*prey[i-1]*predator[i-1])
    predator[i] = predator[i-1] + h * (-c*predator[i-1] + d*prey[i-1]*predator[i-1])

# Plotting
plt.figure()
plt.plot(time, prey, label="Prey Population")
plt.plot(time, predator, label="Predator Population")
plt.xlabel("Time")
plt.ylabel("Population")
plt.title("Predatorâ€“Prey Population Dynamics (Euler Method)")
plt.legend()
plt.grid(True)
plt.show()
```
